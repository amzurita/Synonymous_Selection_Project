### Get the Ne

```python
mu=1.5E-8
Ls=9728061

Na_fromSim={}

for i in syn_theta_data: 
    Na_fromSim_replicate={k:(v/(4*Ls*mu)) for k,v in syn_theta_data[i].items()}
    Na_fromSim[i]=Na_fromSim_replicate
```

```python
Ncurr_data={}

for condition in Na_fromSim:
    Ncurr_data[condition]={}
    for replicate_id in Na_fromSim[condition]: 
        Na=Na_fromSim[condition][replicate_id]
        fold_expansion=dem_parameteres_data[condition][replicate_id][0]
        Ncurr=Na*fold_expansion
        Ncurr_data[condition][replicate_id]=Ncurr
        #print(Ncurr)
```

### Compute the time in generations 

```python
# The time T is in terms of 2*Na. To get generations, multiply T by 2*Na
dem_parameteres_generations_data={}
t_values=[]

for condition in dem_parameteres_data:
    dem_parameteres_generations_data[condition]={}
    
    for replicate_id in dem_parameteres_data[condition]:
        Na=syn_theta_data[condition][replicate_id]
        model=best_model_data[condition][replicate_id]
        
        #For the OneEpoch, there is no time change to be calculated
        if model=='OneEpoch':
            new_nu=dem_parameteres_data[condition][replicate_id][0]
            new_t=0
            dem_parameteres_generations_data[condition][replicate_id]=[new_nu, new_t]
            
        elif model=='TwoEpoch':
            new_nu=dem_parameteres_data[condition][replicate_id][0]
            old_t=dem_parameteres_data[condition][replicate_id][1]
            new_t=old_t*Na*2
            t_values.append(new_t)
            dem_parameteres_generations_data[condition][replicate_id]=[new_nu, new_t]
            

max_t_value=max(t_values)
max(t_values)
```

### Generate Residuals

```python
#Generate a residuals dataframe by adding the columns R-range to the existing dataframes
s_dhet_range_names=['0toE5','E5toE4', 'E4toE3', 'E3toE2', 'MoreThanE2']
gamma_range_names=['0-0.1', '0.1-1', '1-10','10-100','>100']

#Compute the residuals
for i in range(5):
    true_sdhet_discrete=s_dhet_discretized_true[i]
    true_gamma_discrete=gamma_discretized_true[i]
    
    name_sdhet=s_dhet_range_names[i]
    name_gamma=gamma_range_names[i]
    
    name_sdhet_residual='R'+name_sdhet
    name_gamma_residual='R'+name_gamma
    
    s_dhet_dataframe[name_sdhet_residual]=(s_dhet_dataframe[name_sdhet]-true_sdhet_discrete)/np.sqrt(true_sdhet_discrete)
    gamma_dataframe[name_gamma_residual]=(gamma_dataframe[name_gamma]-true_gamma_discrete)/np.sqrt(true_gamma_discrete)
```
