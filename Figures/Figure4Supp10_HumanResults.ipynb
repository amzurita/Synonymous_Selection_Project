## Compute time in generations

```python
# The time T is in terms of 2*Na. To get generations, multiply T by 2*Na
dem_parameteres_generations_data={}
t_values=[]

for condition in dem_parameteres_data:
    dem_parameteres_generations_data[condition]={}
    
    for replicate_id in dem_parameteres_data[condition]:
        Na=syn_theta_data[condition][replicate_id]
        model=best_model_data[condition][replicate_id]
        
        #For the OneEpoch, there is no time change to be calculated
        if model=='OneEpoch':
            new_nu=dem_parameteres_data[condition][replicate_id][0]
            new_t=0
            dem_parameteres_generations_data[condition][replicate_id]=[new_nu, new_t]
            
        elif model=='TwoEpoch':
            new_nu=dem_parameteres_data[condition][replicate_id][0]
            old_t=dem_parameteres_data[condition][replicate_id][1]
            new_t=old_t*Na*2
            t_values.append(new_t)
            dem_parameteres_generations_data[condition][replicate_id]=[new_nu, new_t]
            

max_t_value=max(t_values)
max(t_values)
```

```python
#Build a dataframe for the data for plotting
df_rows=[] #row of the format condition, type(either Na or Ncurr), replicate_id, N_value

#Make the rows in terms of a real value
t_ancestral=-1*max_t_value-15 #Adding 15 to set an "ancestral" time

for condition in Na_fromSim:
    new_rows=[[condition, t_ancestral, replicate_id, Nvalue, best_model_data[condition][replicate_id]] for replicate_id,Nvalue in Na_fromSim[condition].items()]
    df_rows=df_rows+new_rows
    
    new_rows=[[condition, -1*dem_parameteres_generations_data[condition][replicate_id][1],replicate_id, Nvalue, best_model_data[condition][replicate_id]] for replicate_id,Nvalue in Na_fromSim[condition].items()]
    df_rows=df_rows+new_rows

for condition in Ncurr_data: 
    new_rows=[[condition, 0, replicate_id, Nvalue, best_model_data[condition][replicate_id]] for replicate_id,Nvalue in Ncurr_data[condition].items()]
    df_rows=df_rows+new_rows
    
    new_rows=[[condition, -1*dem_parameteres_generations_data[condition][replicate_id][1],replicate_id, Nvalue, best_model_data[condition][replicate_id]] for replicate_id,Nvalue in Ncurr_data[condition].items()]
    df_rows=df_rows+new_rows
    
N_dataframe=pd.DataFrame(df_rows, columns=['Condition', 'Timepoint', 'Replicate_id', 'N_Value', 'Model'])
```
