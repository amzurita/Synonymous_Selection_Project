## Turn DFE Parameters into a dataframe

```python
#Build a dataframe for the data for plotting
dfe_dataframe=[] #row of the format condition, type(either Na or Ncurr), replicate_id, N_value

for condition in dfe_parameters_data:   
    new_rows=[[condition, replicate_id, dfe_parameters_data[condition][replicate_id][0], dfe_parameters_data[condition][replicate_id][1], syn_theta_data[condition][replicate_id]] for replicate_id in dfe_parameters_data[condition].keys()]
    dfe_dataframe=dfe_dataframe+new_rows

DFE_dataframe=pd.DataFrame(dfe_dataframe, columns=['Condition','Replicate_ID', 'Shape_Param', 'Scale_Param', 'Syn_Theta'])
```

```python
#Obtain the Na and add it to the dataframe
mu=1.5E-8
Ls=9728061

DFE_dataframe['Na']=DFE_dataframe['Syn_Theta']/(4*Ls*mu)
```

```python
#Put the Scale_parameter in terms of s_dhet
#Pass the scale parameter to s_dhet, by doing s_dhet=b/(2*N_a)

DFE_dataframe['s_dhet']=DFE_dataframe['Scale_Param']/(2*DFE_dataframe['Na'])
```

## Sample a gamma distribution to get a discretized DFE

```python
order_plotting=['Neutral', 'E5', 'E4','E5_Partial', 'E4_Partial', 'E3_Partial'] #'E3'
```

```python
#For each condition, generate random draws from a gamma distribution. First, in terms of s_dhet
sdhet_histogram_rows=[]

for row in DFE_dataframe.itertuples():
    a=row.Shape_Param
    b=row.s_dhet
    
    gamma_draws=stats.gamma.rvs(a, scale=b,size=10000)
    counts,edges=np.histogram(gamma_draws,bins=[0,1E-5,1E-4,1E-3,1E-2,800000000000])
    #print(row.Replicate_ID, row.Condition)
    #print(max(gamma_draws))
    norm_counts=counts/10000
    new_row=np.append(norm_counts, [row.Condition,row.Replicate_ID])
    sdhet_histogram_rows.append(new_row)

s_dhet_dataframe=pd.DataFrame(sdhet_histogram_rows, columns=['0toE5','E5toE4', 'E4toE3', 'E3toE2', 'MoreThanE2', 'Condition', 'Replicate_ID'])

for i in ['0toE5','E5toE4', 'E4toE3', 'E3toE2', 'MoreThanE2']:
    s_dhet_dataframe[i]=pd.to_numeric(s_dhet_dataframe[i])
```

```python
#For each condition, generate random draws from a gamma distribution. First, in terms of scale_gamma
gamma_histogram_rows=[]

for row in DFE_dataframe.itertuples():
    a=row.Shape_Param
    b=row.Scale_Param
    gamma_draws=stats.gamma.rvs(a, scale=b,size=10000)
    counts,edges=np.histogram(gamma_draws,bins=[0,0.1,1,10,100,100000000000])
    norm_counts=counts/10000
    #print(row.Replicate_ID, row.Condition)
    #print(max(gamma_draws))
    new_row=np.append(norm_counts, [row.Condition,row.Replicate_ID])
    gamma_histogram_rows.append(new_row)

gamma_dataframe=pd.DataFrame(gamma_histogram_rows, columns=['0-0.1', '0.1-1', '1-10','10-100','>100', 'Condition', 'Replicate_ID'])

for i in ['0-0.1', '0.1-1', '1-10','10-100','>100']:
    gamma_dataframe[i]=pd.to_numeric(gamma_dataframe[i])
```

### Generate Residuals

```python
#Generate a residuals dataframe by adding the columns R-range to the existing dataframes
s_dhet_range_names=['0toE5','E5toE4', 'E4toE3', 'E3toE2', 'MoreThanE2']
gamma_range_names=['0-0.1', '0.1-1', '1-10','10-100','>100']

#Compute the residuals
for i in range(5):
    true_sdhet_discrete=s_dhet_discretized_true[i]
    true_gamma_discrete=gamma_discretized_true[i]
    
    name_sdhet=s_dhet_range_names[i]
    name_gamma=gamma_range_names[i]
    
    name_sdhet_residual='R'+name_sdhet
    name_gamma_residual='R'+name_gamma
    
    s_dhet_dataframe[name_sdhet_residual]=(s_dhet_dataframe[name_sdhet]-true_sdhet_discrete)/np.sqrt(true_sdhet_discrete)
    gamma_dataframe[name_gamma_residual]=(gamma_dataframe[name_gamma]-true_gamma_discrete)/np.sqrt(true_gamma_discrete)
```
