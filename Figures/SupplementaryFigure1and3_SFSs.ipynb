## Functions

```python
def load_sfs(filepath, condition, sufix_file='_Replicate1.txt'):
    """
    Opens the SFS file and returns a list of ints. Ignores 0-tons. Assumes file path is filepath+condition+sufix_file. 
    """
    file_name=filepath+condition+sufix_file

    with open(file_name, 'r') as file:
        lines=file.readlines()

    SFS_list=[int(x) for x in lines[1].strip().split()]
    clean_SFS=SFS_list[1:]
    
    return clean_SFS
```

## SFSs side to side

```python
NonSyn_SFS_Summed={}
Syn_SFS_Summed={}

for condition in condition_names:
    values_nonsyn=np.array(load_sfs(folder_path_NonSyn, condition))
    values_syn=np.array(load_sfs(folder_path_Syn, condition))
    
    values_summed_syn=np.append(values_syn[0:9],[values_syn[9:].sum()])
    values_summed_nonsyn=np.append(values_nonsyn[0:9],[values_nonsyn[9:].sum()])
    NonSyn_SFS_Summed[condition]=values_summed_nonsyn
    Syn_SFS_Summed[condition]=values_summed_syn
```

```python
fig, axs = plt.subplots(1, 2, sharex=True, sharey=True, figsize=[20,5])


#Plot the Synonymous first
ax=axs[0]
plt.rcParams['font.size'] = '17'

bins=len(Syn_SFS_Summed['0'])
categories=len(condition_names)

width = 0.25
totalwidthbin=width*(categories+1)
ind = np.array([0+(totalwidthbin*n) for n in range(bins)])
spacing_width=0.1

count=0
for k in condition_names:
    x_pos=ind+(width*count)#+spacing[k]
    bar1 = ax.bar(x_pos, Syn_SFS_Summed[k], width, align='edge', color = color_conditions[k], label=k)
    count=count+1

ax.text(0.05,12500,'Synonymous')
    
#Tick setting
x_labels=x_ticks=np.append(np.arange(1,10).astype(str),[">10"])
x_tickpos=[x+(width*3) for x in ind]
ax.set_xticks(x_tickpos)
ax.set_xticklabels(x_labels)


#Plot the Non-Synonymous second
ax=axs[1]
plt.rcParams['font.size'] = '17'

bins=len(NonSyn_SFS_Summed['0'])
categories=len(condition_names)

width = 0.25
totalwidthbin=width*(categories+1)
ind = np.array([0+(totalwidthbin*n) for n in range(bins)])
spacing_width=0.1

count=0
for k in condition_names:
    x_pos=ind+(width*count)#+spacing[k]
    bar1 = plt.bar(x_pos, NonSyn_SFS_Summed[k], width, align='edge', color = color_conditions[k], label=k)
    count=count+1

x_labels=x_ticks=np.append(np.arange(1,10).astype(str),[">10"])
x_tickpos=[x+(width*3) for x in ind]

ax.text(0.05,12500,'Non-Synonymous')

ax.set_xticks(x_tickpos)
ax.set_xticklabels(x_labels)
fig.supxlabel("# of chromosomes with SNP (frequency)", y=-0.05)
fig.supylabel("Number of SNPs", x=0.06)
plt.savefig("/u/project/klohmuel/amzurita/Synonimous_Selection_Simulations/FigurePlots_Revision/SuppFig3_SFSComparisons.jpg", bbox_inches='tight', dpi=300)
```
